```python
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator

plt.style.use('seaborn-v0_8')
```

# 1) Show f(x) is NOT differentiable at x=3 (kink) and differentiable at x=0

```python
def g(x):
    x = np.asarray(x)
    return 0.5 * x**3 - 1.5 * x**2
    
def f(x):
    x = np.asarray(x)
    return np.abs(g(x)) + 0.5 * x
  
def prime_numerical(g,x):        
    eps = np.finfo(np.float64).eps
    hs = np.sqrt(eps) * max(1.0, abs(x))  # heuristic for central diff- balances truncation and rounding errors
    return (np.abs(g(x+hs)) - np.abs(g(x-hs))) / (2*hs)

def prime_numerical_left(g,x):        
    eps = np.finfo(np.float64).eps
    hs = np.sqrt(eps) * max(1.0, abs(x))
    return (np.abs(g(x)) - np.abs(g(x-hs))) / hs

def prime_numerical_right(g,x):        
    eps = np.finfo(np.float64).eps
    hs = np.sqrt(eps) * max(1.0, abs(x))
    return (np.abs(g(x+hs)) - np.abs(g(x))) / hs
    
def g_prime_analytical(g,x):
    if x==0:
        return (1.5*x**2 - 3*x)
    return np.sign(g(x))*(1.5*x**2 - 3*x)
    
g_rows = []
f_rows = []

for x in [-1.,0.,1.0,3.0]:
    g_val = np.abs(g(x))
    gpa = g_prime_analytical(g,x)
    gpn = prime_numerical(g,x)
    gpnl = prime_numerical_left(g,x)
    gpnr = prime_numerical_right(g,x)
    err = gpn - gpa
    
    g_rows.append({
        "x": x,
        "g(x)": f"{g_val:.3f}",
        "analytical g'(x)": f"{gpa:.3f}",
        "numerical g'(x) central": f"{gpn:.3f}",
        "numerical g'(x) left": f"{gpnl:.3f}",
        "numerical g'(x) right": f"{gpnr:.3f}",
        "error(analytical-central)": f"{err:.0e}",
        "differentiable": f"{int(gpnl - gpnr) == 0 }",
    })

    f_val = np.abs(f(x))    
    fpnl = prime_numerical_left(f,x)
    fpnr = prime_numerical_right(f,x)
    
    f_rows.append({
        "x": x,
        "f(x)": f"{f_val:.3f}",        
        "numerical f'(x) left": f"{fpnl:.3f}",
        "numerical f'(x) right": f"{fpnr:.3f}",        
        "differentiable": f"{int(fpnl - fpnr) == 0 }",
    })

gdf = pd.DataFrame(g_rows)
print(gdf)

print("\ng(x) is NOT differentiable at 3 as left and right slopes are different but differentiable at 0")

fdf = pd.DataFrame(f_rows)
print(fdf)

print("\nf(x) is NOT differentiable at 3 as left and right slopes are different but differentiable at 0")
```

# 2) Analyze the function with plots

```python
def plot_non_differential_points(f, x, non_differential_x_points):
    fig, ax = plt.subplots(figsize=(6.8, 3.6), dpi=90)
    ax.plot(x, f(x), color='C3', lw=1.6, label=f"{f.__name__}(x) - with NDEs")
    ax.xaxis.set_major_locator(MultipleLocator(0.5))
    ax.set_xlabel('x')
    ax.set_ylabel(f"{f.__name__}(x)")
    
    for i, x_val in enumerate(non_differential_x_points):        
        ax.plot(x_val, f(x_val), marker='o', color='orange', markersize=8, label="NDE")  
        
    fig.tight_layout()
    fig.canvas.manager.set_window_title(f"{f.__name__}(x) Visualisation")
    ax.legend()
    plt.show()

def plot(f, x, minmax_candidate_x_points=None, inflection_x_points=None,highlight_0_y=False):
    fig, ax = plt.subplots(figsize=(6.8, 3.6), dpi=90)
    ax.plot(x, f(x), color='C3', lw=1.6, label=f"{f.__name__}(x)")
    ax.xaxis.set_major_locator(MultipleLocator(0.5))
    ax.set_xlabel('x')
    ax.set_ylabel(f"{f.__name__}(x)")
    
    if highlight_0_y:
        ax.axhline(y=0, color='grey', linestyle='--')
    
    if minmax_candidate_x_points != None:        
        for x_val in minmax_candidate_x_points:
            # vertical line
            ax.axvline(x=x_val, color='grey', linestyle='--', linewidth=2)
        
            # arrow showing slope direction at that x
            # choose a small step to the right for the arrow head
            x0 = x_val
            x1 = x_val + 0.2
        
            ax.annotate(
                "",
                xy=(x1, f(x1)),      # arrow head
                xytext=(x0, f(x0)),  # arrow tail
                arrowprops=dict(arrowstyle="->", lw=2, color="black")
            )

            #mark the highlighted
            ax.plot(x_val, f(x_val), 'bo', markersize=8)

    if inflection_x_points != None:        
        for x_val in inflection_x_points:
            ax.axvline(x=x_val, color='green', linestyle=':', linewidth=1, label="Inflection")
            
            #mark the highlighted
            ax.plot(x_val, f(x_val), marker='*', color='green', markersize=8)
            
    fig.tight_layout()
    fig.canvas.manager.set_window_title(f"{f.__name__}(x) Visualisation")
    ax.legend()
    plt.show()

# coefficients of: ax^2 + bx + c = 0
def get_roots(a, b, c):
    roots = np.roots([a, b, c])

    if a == 0:
        # Linear case: bx + c = 0
        return roots[0], None
        
    # Quadratic case
    x_minus, x_plus = np.sort(roots)
    return x_minus, x_plus

# f(x) = |g(x)| + 0.5x is piecewise differentiable:
# If g(x) < 0:  f'(x) = -(g'(x)) + 0.5
# If g(x) > 0:  f'(x) =  (g'(x)) + 0.5
def df(x):
    x = np.asarray(x)

    # g'(x)
    gp = 1.5*x**2 - 3*x

    # sign(g(x)) is -1 for x<3, +1 for x>3, x=0 sign is irrevalent as g(0)=0
    sign = np.where(x < 3, -1.0, 1.0)

    # derivative of |g(x)| is s * g'(x), except at x=3
    # but f'(x) exists everywhere because of +0.5*x
    abs_gp = sign * gp

    # override undefined |g(3)|
    abs_gp = np.where(x == 3, 0.0, abs_gp)

    return abs_gp + 0.5 #@TODO: recheck if is this correct for x=3 as g(3) is undefined
    
def d2f(x):
    x = np.asarray(x)
    f2 = 3*x - 3
    sign = np.where(x < 3, -1.0, 1.0)
    abs_f2 = sign * f2

    # f''(3) does not exist â†’ return np.nan
    out = np.where(x == 3, np.nan, abs_f2)
    
    return abs_f2


# utility function
def bold(text: str) -> str:
    return f"\033[1m{text}\033[0m"

print(f"Potential points of non-differentiability in f(x) are g(0)={g(0)} and g(3)={g(3)}")

slope_at_0 = df(0)

try:
    slope_at_3 = df(3)
except ValueError as e:
    slope_at_3=str(e)

print(f"f(x) is differentiable at critical point x=0 f'(x)={slope_at_0}")
print(f"f(x) is nonâ€‘differentiable at x=3 because g(3)=0 and the absolute value |g(x)| causes the left({-1*3*(3-1)}) and right({3*(3-1)}) slopes to differ")
print("\nPiece wise regions are: (-infinity,3) and(3,+infinity). g(x) < 0 in (-infinity,3) region and > 0 in (3, +infinity)")

x = np.linspace(-2.5, 4, 400)
plot_non_differential_points(f, x, [3])

# g(x) derivative has 2 definitions due to abs
# g(x) < 0 => g'(x) = -1 * (1.5 * x * (x-2))
# g(x) > 0 => g'(x) = (1.5 * x * (x-2))
# roots of the derivative for the negative branch (where g(x) < 0 ie (-infinity,3) region)
# first_derivative(-ve) roots: -0.1547 2.1547
(x_minus, x_plus) = get_roots(-1.5,3.,0.5)

possible_minmax_x_points = set()

# only x in region (-infinity,3) are valid roots, otherwise there are no solutions for the +ve side
if x_minus < 3:
    print('\ng(x)<0 region:')
    print(f" slope is {df(x_minus):.4f} at ({x_minus:.4f},{f(x_minus):.4f}) and curvature is {'+ve' if d2f(x_minus) > 0 else '-ve'}")
    possible_minmax_x_points.add(x_minus)
if x_plus < 3:
    print(f" slope is {df(x_plus):.4f} at ({x_plus:.4f},{f(x_plus):.4f}) and curvature is {'+ve' if d2f(x_plus) > 0 else '-ve'}")
    possible_minmax_x_points.add(x_plus)
   
# roots of the derivative for the positive branch (where g(x) > 0 ie (3,+infinity) region)
# first_derivative(+ve) roots: -1.8165 -0.1835
(x_minus, x_plus) = get_roots(1.5,-3.,0.5)

# only x in region (3,+infinity) are valid roots, otherwise there are no solutions for the +ve side
if x_minus > 3:
    print('g(x)>0 region:')
    print(f" slope is {df(x_minus):.4f} at ({x_minus:.4f},{f(x_minus):.4f}) and curvature is {'+ve' if d2f(x_minus) > 0 else '-ve'}")
    possible_minmax_x_points.add(x_minus)
if x_plus > 3:
    print(f" slope is {df(x_plus):.4f} at ({x_plus:.4f},{f(x_plus):.4f}) and curvature is {'+ve' if d2f(x_plus) > 0 else '-ve'}")
    possible_minmax_x_points.add(x_plus)

inflection_x_points=set()
(x_inf,_) = get_roots(0.,-3.,3.)
inflection_x_points.add(x_inf)
(x_inf,_) = get_roots(0.,3.,-3.)
inflection_x_points.add(x_inf)

print(f"\nInflection points are:")
for x_inf in inflection_x_points:
    print(
        f"\tx={x_inf:.4f} curvature={d2f(x_inf):.4f}. "
        f"The curvature from left({d2f(x_inf-2.0)}) "
        f"to right({d2f(x_inf+1.0)}) around this point is "
        f"{bold('upwards') if d2f(x_inf-1.0) < 0 
          and second_derivative(x_inf+1.0) > 0 
          else bold('downwards')}"
    )
    
plot(f, x, minmax_candidate_x_points=possible_minmax_x_points, inflection_x_points=inflection_x_points)
plot(df, x, minmax_candidate_x_points=possible_minmax_x_points, highlight_0_y=True)
plot(d2f, x, minmax_candidate_x_points=possible_minmax_x_points, inflection_x_points=inflection_x_points, highlight_0_y=True)
print(f"\nBased on the facts that \n\t1) f(x) -> +ve infinity when x-> +-infinity"
      f"\n\t2) -0.1547 is the only local min point as the curvature at this point is +ve"
      f"\n {bold('(-0.1547,-0.0396) is the global minimum')}")
```

# 3) Gradient descent(GD)

```python
def gradient_step(x, eta, delta):
    return x - eta*delta
    
def plot_gd_single_step(f, x, eta, starting_positions, new_positions, global_min):
    fig, ax = plt.subplots(figsize=(8.8, 3.6), dpi=90)
    ax.plot(x, f(x), color='C3', lw=1.6, label="f(x)")
    ax.xaxis.set_major_locator(MultipleLocator(0.5))
    ax.set_xlabel('x')
    ax.set_ylabel(f"{f.__name__}(x)")
    
    colors=['blue','purple','green','black','brown']
    for i, x_val in enumerate(starting_positions): 
        c = colors[i % len(colors)]
        ax.plot(x_val, f(x_val), marker='*', color=c, markersize=8, label='Starting position')  

    for i, x_val in enumerate(new_positions): 
        c = colors[i % len(colors)]
        ax.plot(x_val, f(x_val), marker='o', color=c, markersize=8, label='New position')  

    ax.axvline(x=global_min, color='red', linestyle='--', linewidth=2)
    
    fig.tight_layout()
    fig.canvas.manager.set_window_title(f"{f.__name__}(x) GD Steps")
    ax.set_title(f"GD steps(single) - different starting points(eta={eta})")
    ax.legend()
    plt.show()
    
starting_positions=[-3.5,-1,0,1,3.6]
eta=0.15
delta=df(starting_positions)
new_positions=gradient_step(starting_positions, eta, delta)
x = np.linspace(-5, 4, 400)
plot_gd_single_step(f,x,eta,starting_positions, new_positions,global_min=-0.1547)
print("The GD step moves towards the global min from either side if starting point is from a low curvature area, else points like -3.5 overshoots\n")

def gradient_descent(df, x0, eta, steps):
    xs = [np.array(x0, dtype=np.float64)]
    for _ in range(steps):
        x = xs[-1][0] - eta * df(xs[-1][0])
        y = f(x)
        xs.append([x,y])
    return np.stack(xs)

# plot the GD steps
start = np.array([1.0, f(1.0)])
slow = gradient_descent(df, start, eta=0.08, steps=40)
fast = gradient_descent(df, start, eta=0.18, steps=40)

print(f"Gloal min: [-0.15470053837925152,{f(-0.15470053837925152)}]")
print(f"Last GD after 40 steps reached in slow(eta=0.08): x={slow[-1]}")
print(f"Last GD after 40 steps reached in fast(eta=0.18): x={fast[-1]}\n")

fig, ax = plt.subplots(figsize=(6.4, 3.6), dpi=90)
levels = np.linspace(0.0, 6.0, 20)
xs = np.linspace(-2.0, 2.0, 200)
ys = np.linspace(-2.0, 2.0, 200)
X, Y = np.meshgrid(xs, ys)
Z = 0.5 * (10 * X**2 + Y**2)
ax.contour(X, Y, Z, levels=levels, cmap="Blues")
ax.plot(slow[:, 0], slow[:, 1], "s-", ms=4, color="gold", label="eta=0.08")
ax.plot(fast[:, 0], fast[:, 1], "o-", ms=2, color="brown", label="eta=0.18")

ax.plot(-0.15470053837925152, f(-0.15470053837925152), marker='o', color='red', markersize=8, label='Global Min') 

ax.legend(loc="upper right")
ax.set_xlabel("x")
ax.set_ylabel("y")
ax.set_title("Gradient descent trajectories")
plt.show()



```

```python
def plot_gd_steps(f, x, gd_steps_low_eta=None, gd_steps_high_eta=None, global_min=-0.15470053837925152, title=None, high_eta_label="High Eta", low_eta_label="Low Eta"):
    fig, ax = plt.subplots(figsize=(8.8, 3.6), dpi=90)
    ax.plot(x, f(x), color='C3', lw=1.6, label="f(x)")
    ax.xaxis.set_major_locator(MultipleLocator(0.5))
    ax.set_xlabel('x')
    ax.set_ylabel(f"{f.__name__}(x)")

    if gd_steps_high_eta is not None:
        ax.plot(gd_steps_high_eta[:, 0], gd_steps_high_eta[:, 1], color='brown',marker='o', lw=1.6, label=high_eta_label)
        
        for (x0, y0), (x1, y1) in zip(gd_steps_high_eta[:-1], gd_steps_high_eta[1:]):
            xm = (x0 + x1) / 2
            ym = (y0 + y1) / 2
        
            # small offset so the arrow has a direction
            dx = (x1 - x0) * 0.1
            dy = (y1 - y0) * 0.1
    
            ax.annotate(
                "",
                xy=(xm, ym),      # arrow head
                xytext=(xm - dx, ym - dy),  # arrow tail
                arrowprops=dict(arrowstyle="-|>,head_length=1,head_width=.5", color="black", lw=.6)
            )
    if gd_steps_low_eta is not None:
        ax.plot(gd_steps_low_eta[:, 0], gd_steps_low_eta[:, 1], color='red', lw=1.6, label=low_eta_label)
    
    ax.axvline(x=global_min, color='green', linestyle='--', linewidth=2)
    ax.plot(-0.15470053837925152, f(-0.15470053837925152), marker='o', color='orange', markersize=8, label='Global Min') 
    
    fig.tight_layout()
    fig.canvas.manager.set_window_title(f"{f.__name__}(x) GD Steps")
    ax.legend()
    ax.set_title(title)
    plt.show()

x = np.linspace(-1.5, 2, 400)
start = np.array([.2, f(.2)])
low_eta = gradient_descent(df, start, eta=0.08, steps=5)
high_eta = gradient_descent(df, start, eta=0.9, steps=5)
plot_gd_steps(f,x, gd_steps_low_eta=low_eta, gd_steps_high_eta=high_eta, title="Gradient descent trajectories - Overshooting and oscillation", high_eta_label="GD (Î· = 0.9)",
             low_eta_label="GD (Î· = 0..08)")


def gd_backtrack(f, df, x0, eta=1e-1, beta=0.5, c=1e-4, steps=40):    
    xs = [np.array(x0, dtype=np.float64)]
    
    for _ in range(steps):
        x = xs[-1][0]
        grad = df(x)
        t = eta

        # squared gradient norm
        grad2 = grad*grad

        fx = f(x)

        # Armijo condition: f(x - t grad) <= f(x) - c t ||grad||^2
        while f(x - t*grad) > fx - c * t * grad2:
            t *= beta

        x = x - t*grad        
        xs.append([x, f(x)])

    return np.stack(xs)
	
# plot the GD steps
start = np.array([1.0, f(1.0)])
gd_backtrack_steps = gd_backtrack(f, df, start, eta=0.9, steps=40)
plot_gd_steps(f,x, gd_steps_high_eta=gd_backtrack_steps, title="Gradient descent trajectory - High Eta with Backtracking, converges quickly",high_eta_label="HB (Î· = 0.9, Î² = 0.5)")

start = np.array([1.5, f(1.5)])
x = np.linspace(-.5, 2, 400)
gd_normal = gradient_descent(df, start, eta=0.15, steps=60)
gd_backtrack_steps = gd_backtrack(f, df, start, eta=0.08, beta=0.9, steps=60)
plot_gd_steps(f,x, gd_steps_low_eta=gd_backtrack_steps, gd_steps_high_eta=gd_normal, title="Gradient descent trajectory - Normal(High Eta) vs Backtracking(Low Eta)",
             high_eta_label="GD (Î· = 0.15)", low_eta_label="HB (Î· = 0.08, Î² = 0.9)")
```

# 4) Stochastic Gradient Descent(SGD)

```python
def exact_variance(eta, mu, sigma): 
    return (eta * sigma**2) / (2.0*mu - eta*mu**2.0)

def approx_variance(eta, mu, sigma):
    return eta * sigma**2 / (2 * mu)

def simulate_sgd_root_mean_square_distance(eta, mu, sigma, errs):
    steps = len(errs)
    ys = [0.0]

    for i,t in enumerate(np.arange(steps-1)):        
        y = (1 - eta * mu) * ys[-1] - eta * errs[i]
        ys.append(y)
        
    return np.array(ys)

def simulate_sgd(eta, mu, errs, x0):
    no_of_steps = len(errs)
    xy = [[x0, f(x0)]]

    for i,t in enumerate(np.arange(no_of_steps-1)):   
        if np.isscalar(eta):
            eta_t = eta
        else:
            eta_t = eta[i]
            
        last_x = xy[-1][0]
        x_new = last_x - (eta_t * (df(last_x) + errs[i]))
        y_new = f(x_new)
        xy.append([x_new,y_new])
        
    return np.array(xy)
    
k=50000
global_min=-0.15470053837925152
x = np.linspace(-1, 4, k)
mu_max=d2f(global_min)
mu=0.7*mu_max # allowed range : (0,mu_max)
eta_max=2.0/mu
print("mu:", mu)
print("eta_max:", eta_max)

etas = np.linspace(eta_max*.01, eta_max*.99, 40) # eta = 2/mu is a boundary condition resulting in undefined behaviour as denominator becomes 0

rng = np.random.default_rng(seed=23)
sigma = 0.001
errs = rng.normal(0,sigma, k)

empirical = []
exact = []
approx = []

for eta in etas:
    ys = simulate_sgd_root_mean_square_distance(eta, mu, sigma, errs)
    empirical.append(np.mean(ys**2))
    exact.append(exact_variance(eta, mu, sigma))
    approx.append(approx_variance(eta, mu, sigma))

fig, ax = plt.subplots(figsize=(10,3))
ax.plot(etas, empirical, 'o-', label="Empirical variance")
ax.plot(etas, exact, 'r--', label="Exact variance")
ax.plot(etas, approx, 'g--', label="Approx variance")
ax.set_xlabel("x")
ax.set_ylabel("Steady state variance")
ax.set_title("Steady State Variance of SGD - Comparision")
ax.legend()
ax.grid(True)
fig.tight_layout()
plt.show()
#os.makedirs("assets", exist_ok=True)
#plt.savefig('assets//steady_state_variance.pdf',format='pdf', dpi=900, bbox_inches="tight")

x0=2.0
k=2500
x = np.linspace(-1, 4, k)
eta=eta_max*.01
rng = np.random.default_rng(seed=2)

errs1 = rng.normal(0, 0.001, k)
errs2 = rng.normal(0, 0.01,  k)
errs3 = rng.normal(0, 0.9,   k)

eta1=eta_max*0.001
eta2=eta_max*0.01
eta3=eta_max*0.1

xy1 = simulate_sgd(eta1, mu, errs1, x0)
xy2 = simulate_sgd(eta2, mu, errs2, x0)
xy3 = simulate_sgd(eta3, mu, errs3, x0)

# Plot SGD

fig, ax = plt.subplots(figsize=(10,4))
ax.plot(xy1[:,0], xy1[:,1], 'o',color='black', label=f"sgd(eta={eta1:.4f}, sigma=0.001)")
ax.plot(xy2[:,0], xy2[:,1], '*',color='red', label=f"sgd(eta={eta2:.4f}, sigma=0.01)")
ax.plot(xy3[:,0], xy3[:,1], '-',color='green', label=f"sgd(eta={eta3:.4f}, sigma=0.9)")
ax.set_ylabel("y")
ax.set_xlabel("x")
ax.set_title("SGD")
ax.plot(global_min, f(global_min), marker='o', color='gold', markersize=8, label='Global Min')
ax.legend()
ax.grid(True)
fig.tight_layout()
plt.show()
#os.makedirs("assets", exist_ok=True)
#plt.savefig('assets//sgd_trajectories.pdf',format='pdf', dpi=900, bbox_inches="tight")

print("Small learning rate(eta) and small noise(sigma) give slow but stable convergence and a tight cluster around the minimum." 
    + "Increasing both eta and sigma speeds up early progress but leads to larger fluctuations around the optimum. With large eta and sigma, "
    + "the trajectory becomes highly volatile and may fail to settle")


# Plot x vs k
steps1 = np.arange(len(xy1))
steps2 = np.arange(len(xy2))
steps3 = np.arange(len(xy3))

fig, ax = plt.subplots(figsize=(10,4))
ax.plot(steps1, xy1[:,0], label=f"eta={eta1:.4f} trajectory", color="black")
ax.plot(steps2, xy2[:,0], label=f"eta={eta2:.4f} trajectory", color="red")
ax.plot(steps3, xy3[:,0], label=f"eta={eta3:.4f} trajectory", color="green")
ax.axhline(f(global_min), color='gold', linestyle='--', linewidth=2, label='y at global min')
ax.set_xlabel("iteration k")
ax.set_ylabel("x value")
ax.set_title("SGD: x vs Iterations")
ax.grid(True)
ax.legend()

fig.tight_layout()
plt.show()
#os.makedirs("assets", exist_ok=True)
#plt.savefig('assets//sgd_x_vs_iterations.pdf',format='pdf', dpi=900, bbox_inches="tight")

# Plot distance from the minimum over time
dist1 = xy1[:,0] - global_min
dist2 = xy2[:,0] - global_min
dist3 = xy3[:,0] - global_min

fig, ax = plt.subplots(figsize=(10,4))
ax.plot(dist1, label=f"eta={eta1:.4f}", color="black")
ax.plot(dist2, label=f"eta={eta1:.4f}", color="red")
ax.plot(dist3, label=f"eta={eta1:.4f}", color="green")

ax.axhline(0.0, color="gold", linestyle="--", label="at minimum")
ax.set_xlabel("Iteration k")
ax.set_ylabel("Distance from minimum")
ax.legend()
ax.set_title("SGD: Distance from minimum")
ax.grid(True)
plt.show()
#os.makedirs("assets", exist_ok=True)
#plt.savefig('assets//sgd_distance_from_min.pdf',format='pdf', dpi=900, bbox_inches="tight")


# Plot constant vs reducing learning rate(eta)
k_decaying_eta=4000
errs_decaying_eta = rng.normal(0, 0.01,  k_decaying_eta)
decaying_eta = [eta2 / (1 + 0.01 * i) for i in range(k_decaying_eta)]
xy_decaying_eta = simulate_sgd(decaying_eta, mu, errs_decaying_eta, x0)
dist_decaying_eta = xy_decaying_eta[:,0] - global_min

fig, ax = plt.subplots(figsize=(10,4))
#ax.xaxis.set_major_locator(MultipleLocator(1000))
ax.plot(dist_decaying_eta, label=f"decaying eta k={k_decaying_eta}", color="black")
ax.plot(dist2, label=f"eta={eta1:.4f} k={k}", color="red")

ax.axhline(0.0, color="gold", linestyle="--", label="at minimum")
ax.set_xlabel("Iteration k")
ax.set_ylabel("Distance from minimum")
ax.legend()
ax.set_title("SGD: Distance from min - Constant Vs Decaying Learning Rate")
ax.grid(True)
plt.show()
#os.makedirs("assets", exist_ok=True)
#plt.savefig('assets//sgd_distance_from_min_cons_vs_decaying.pdf',format='pdf', dpi=900, bbox_inches="tight")
```
